```python
import numpy as np
```

```python
rs = np.random.RandomState(321) # create a random state object
arr = rs.rand(5,5)
print(arr)
```

### sort an array along a specified axis

```python
# sort along the column and return a copy
print(np.sort(arr, axis=0))   
print(" ")
print(arr)  # original array remains unsorted
```

```python
# sort along the column in place
arr.sort(axis=0)
print(arr) # now sorted along the column
```

```python
# sort along the column and return a copy
print(np.sort(arr, axis=1))    
```

```python
# sort along the row in place
arr.sort(axis=1)    
print(arr)
```

### compute the indices that would sort an array along a specified axis

```python
rs = np.random.RandomState(321) # create a random state object
arr = rs.rand(5,5)
print(arr)
```

```python
# along the column - return the indices that would sort the array
print(np.argsort(arr, axis=0))
```

```python
# along the row - return the indices that would sort the array
print(np.argsort(arr, axis=1))
```

```python
# if axis=None, return the indices of a flattened array
print(np.argsort(arr, axis=None))
```
